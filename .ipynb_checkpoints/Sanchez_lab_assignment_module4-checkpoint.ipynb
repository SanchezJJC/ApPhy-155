**The Lotka-Volterra equations**

_Refer to Exercise 8.2 in the Newman text._ The Lotka-Volterra equations are a mathematical model of predator-prey interactions between biological species. Let two variables $x$ and $y$ be proportional to the size of the populations of two species, traditionally called "rabbits" (the prey) and "foxes" (the predators). You could think of $x$ and $y$ as being the population in thousands, say, so that $x = 2$ means there are 2000 rabbits.

In the Lotka-Volterra model the rabbits reproduce at a rate proportional to their population, but are eaten by the foxes at a rate proportional to both their own population and the population of foxes:

$$\frac{{\rm d}x}{{\rm d}t} = \alpha x - \beta xy,$$

where $\alpha$ and $\beta$ are constants. At the same time the foxes reproduce at a rate proportional to the rate at which they eat rabbits-- because they need food to grow and reproduce-- but also die of old age at a rate proportional to their own population:

$$\frac{{\rm d}y}{{\rm d}t}  = \gamma xy - \delta y,$$

where $\gamma$ and $\delta$ are also constants. *(25 pts.)*

1. Solve these equations using the fourth-order Runge-Kutta method for the case $\alpha = 1, \beta = \gamma = 0.5$, and $\delta = 2$, starting from the initial condition $x = y = 2$.

2. Make a graph showing both $x$ and $y$ as a function of time on the same axes from $t =0$ to $t = 30$. *(25 pts.)*

```python
# Importing necessary packages
import numpy as np
import matplotlib.pyplot as plt

# Defining vector for 1st derivatives f(r,t). Here, we also define the constants
#associated with the ODE's
def f(r, t, alpha=1, beta=0.5, gamma=0.5, delta=2):

  # The parameters need to be positive for the differential equations to make sense
  #in the context of prey-predator population dynamics
  if alpha <= 0 or beta <= 0 or gamma <= 0 or delta <= 0:
    return print("alpha, beta, gamma and delta parameters must all be positive")

  # We let the first element of our 'position vector' be x and the second one be y
  x = r[0]
  y = r[1]

  # Here, we define f_x and f_y, the first, which are the first deriviatives of
  #x(t) and y(t) in terms of x, y and t. We input the appropriate Lotka-Voltera
  #equations here
  f_x = alpha*x - beta*x*y
  f_y = gamma*x*y - delta*y

  return np.array([f_x, f_y], float)

# Defining the function that uses 4th-order Runge-Kutta Method
def RK_4(x_initial, y_initial, t_initial=0, t_final=30, precision=1000, \
         use_t_array=False, use_x_array=False, use_y_array=False, use_print=False, **kwargs):
  """
  Solves 2D ODE's using the 4th-order Runge-Kutta Method.

  t_initial and t_final determine the intiial and final time values respectively,
  while the precision parameter determines how many time points are going to be
  used.

  x_intitial and y_initial determine the initial conditions of x and y
  respectively.


  """

  # Initializing time array, x values array and y values array
  # x_array and y_array are lists. This allows us to use the method .append() later
  t_array = np.linspace(start=t_initial, stop=t_final, num=precision)
  x_array = []
  y_array = []

  # If the user sets more than one use_..._array parameter True, the parameter
  #higher up the list takes priority. We put this if statement here so that
  #the whole program doesn't have to run before returning t_array since it doesn't change.
  if use_t_array == True:
   return t_array

  # Defining the time step h
  h = (t_final - t_initial)/(precision)

  # Initializing position vector (in this case, it refers to the phase space of x and y)
  r_array = np.array([x_initial, y_initial], float)

  # This is the main loop that implements the 4th-order Runge-Kutta Method
  for i in range(len(t_array)): # We run this for every element in t_array

    # The loop starts by appending the initial conditions to the x and y lists
    x_array.append(r_array[0])
    y_array.append(r_array[1])

    # Defining each k vector. We consider t in here even if f(x,y,t) doesn't
    #explicitly contain t for the sake of generality
    k1 = h*f(r=r_array, t=t_array[i])
    k2 = h*f(r=r_array+0.5*k1, t=t_array[i]+0.5*h)
    k3 = h*f(r=r_array+0.5*k2, t=t_array[i]+0.5*h)
    k4 = h*f(r=r_array+k3, t=t_array[i]+h)

    r_array += (1/6)*(k1 + 2*k2 + 2*k3 + k4)

  # These if statements also control what the function outputs
  if use_print == True:
    print(f"\nInitial x (# of rabbits in thousands):{x_array[0]}\nInitial y (# of foxes in thousands):{y_array[0]}\nInitial time:{t_array[0]}\nFinal time:{t_array[-1]}\nTime step:{h}\nx values:{x_array}\ny values:{y_array}\n")
  else:
   if use_x_array == True:
    return x_array

   if use_y_array == True:
    return y_array

# Output of values
RK_4(x_initial=2, y_initial=2, use_print=True)

# Arrays
x_array = RK_4(x_initial=2, y_initial=2, use_x_array=True)
y_array = RK_4(x_initial=2, y_initial=2, use_y_array=True)
t_array = RK_4(x_initial=2, y_initial=2, use_t_array=True)

print(f"alpha = {1}; beta = {0.5}; gamma = {0.5}; delta = {2}")
print(f"\nMaximum Rabbit population:{max(x_array)} thousand \nMinimum Rabbit population:{min(x_array)} thousand \n")
print(f"Maximum Fox population:{max(y_array)} thousand \nMinimum Fox population:{min(y_array)} thousand \n")
#-------------------------------------------------------------------------------------------------

# PLOTTING

# Initializing subplot
fig1, ax1 = plt.subplots()

# Defining a function that tweaks plot settings. I hope to use this (or a better version) even in other assessments that require plotting
def plot_settings(figure, axes, x,y, xlabel="x", ylabel="y", color="blue", title="", \
                  use_xlimit=False, use_ylimit=False, xlim1=0, xlim2=100, ylim1=-10, ylim2=10, \
                  marker="o", markersize=5, linestyle="--", label="", **kwargs):
  """
  Plots specified arrays.

  xlabel and ylabel sets the labels of the respective axes.
  use_xlimit/use_ylimit toggles whether or not plt.xlim()/plt.ylim would be used, where xlim1/ylim1 and
  xlim2/ylim2 are its parameters.
  """
  axes.plot(x,y, marker=marker, markersize=markersize, linestyle=linestyle, color=color, label=label)

  plt.title(title)
  axes.set_xlabel(xlabel)
  axes.set_ylabel(ylabel)

  # These if statements are for incase we want to change the limits of the graph
  if use_xlimit == True:
   plt.xlim(xlim1, xlim2)
   if use_ylimit == True:
     plt.ylim(ylim1, ylim2)

  if use_ylimit == True:
    plt.ylim(ylim1, ylim2)
    if use_xlimit == True:
      plt.xlim(xlim1, xlim2)

# Plotting the fox population and rabbit population over time curves from t=0 to
#t=30
# We set the other pertinent axis labels only once since subsequent calls of
#plot_settings() with said labels will take on the values of the most recent call
plot_settings(fig1,ax1,t_array,y_array, \
              color="mediumblue", label="Foxes", markersize=3)

plot_settings(fig1,ax1,t_array,x_array, xlabel="time", ylabel="Population in thousands", \
              color="limegreen", label="Rabbits", title="Fig 1. Population over time of foxes and rabbits", markersize=3)

fig1.legend()

# Plotting the fox and rabbit populations phase space (Phase spaces correspond to
#the state of a system. In this case, the state of the system can be determined
#by the rabbit and fox populations at a given time)

# Initializing Phase space plot
fig2, ax2 = plt.subplots()

plot_settings(fig2, ax2, x_array, y_array, \
              color="gold", title="Fig 2. Phase space of fox and rabbit populations (in thousands)", \
              xlabel="Rabbit population", ylabel="Fox population")

plt.show()
#---------------------------------------------------------------------------------------------------------
# Plotting another case where alpha=1, beta=2, gamma=2, delta=1

# Making new function
def f(r, t, alpha=1, beta=1, gamma=1, delta=1):
  if alpha <= 0 or beta <= 0 or gamma <= 0 or delta <= 0:
    return print("alpha, beta, gamma and delta parameters must all be positive")

  x = r[0]
  y = r[1]

  f_x = alpha*x - beta*x*y
  f_y = gamma*x*y - delta*y

  return np.array([f_x, f_y], float)

# Output of values
RK_4(x_initial=2, y_initial=2, use_print=True)

# Arrays
x_array = RK_4(x_initial=2, y_initial=2, use_x_array=True)
y_array = RK_4(x_initial=2, y_initial=2, use_y_array=True)
t_array = RK_4(x_initial=2, y_initial=2, use_t_array=True)

print(f"alpha = {1}; beta = {1}; gamma = {1}; delta = {1}")
print(f"\nMaximum Rabbit population:{max(x_array)} thousand \nMinimum Rabbit population:{min(x_array)} thousand \n")
print(f"Maximum Fox population:{max(y_array)} thousand \nMinimum Fox population:{min(y_array)} thousand \n")

# Plotting time evolution of new rabbit and fox system
fig3, ax3 = plt.subplots()

plot_settings(fig3,ax3,t_array,y_array, \
              color="mediumblue", label="Foxes", markersize=3)

plot_settings(fig3,ax3,t_array,x_array, xlabel="time", ylabel="Population in thousands", \
              color="limegreen", label="Rabbits", title="Fig 3. Population over time of foxes and rabbits", markersize=3)

fig3.legend()

# Plotting new rabbit-fox phase space
fig4, ax4 = plt.subplots()

plot_settings(fig4, ax4, x_array, y_array, \
              color="gold", title="Fig 4. Phase space of fox and rabbit populations (in thousands)", \
              xlabel="Rabbit population", ylabel="Fox population")


plt.show()
```

The coupled fox and rabbit populations can be described as a dynamical system given that it is represented by two differential equations. States of such a system can be described by the phase space whose axes correspond to the populations present.

As Fig 1 and Fig 2 show, the system (or the model of which that we're using) is completely deterministic; the rabbit and fox populations show periodicity, and the fox-rabbit phase space is a closed loop. This means that regardless of our initial conditions, we can expect to see the smooth oscillations we find in Fig 1.

(Reproduction rate, predation rate, and death rate are discussed in greater detail in Some Notes 1)

In Fig 1, we can see that the rabbit population tends to have much higher maximum and minimum values than that of the fox population despite starting at the same population size. This is to be expected given the $\alpha$, $\beta$, $\gamma$ and $\delta$ patameters set:
* $\alpha > \beta $ The rabbit reproduction rate outpaces the death rate due to predation

* $\gamma < \delta $ The fox reproduction rate is outpaced by the death rate due to aging. The fox population stil stably oscillates instead of completely terminating despite this since the only way for the foxes to die out in this model is if there are no rabbits (\\(\frac{{\rm d}y}{{\rm d}t}\\) becomes negative for all \\(y\\))

This is in contrast to Fig 3, where \\(\alpha = \beta = \gamma = \delta = 1\\), the minimum and maximum values are identical.  

Fig 1 follows the predator-prey dynamics one can reasonably expect from an ideal system such as this:
* As the rabbit population begins to thrive, the fox population increases if this greater availability of food can make its reproduction rate outpace its death rate. This isn't the case at the start given that the fox population was decreasing.

* When the rabbit population becomes large enough for the fox reproduction rate to finally outpace its death rate, the fox population reaches its minimum and begins to increase. As the fox population begins to increase, the predation term also begins to increase, but it has yet to outpace the rabbit reproduction rate, the rabbit population is still increasing. This also contributes to the increase of the predation rate.    

* As the rabbit population increases, so too does the fox reproduction rate, the fox population and predation rate. Once the predation rate outpaces the rabbit reproduction rate, the rabbit population reaches its peak and begins to decrease.

* Once a sufficient decrease in the rabbit population and a sufficient increase in the fox population are achieved, the fox reproduction rate will become outpaced by the death rate. This will cause the fox population to reach its peak and start to decrease. Because of this, the predation rate, death rate, and both reproduction rates will decrease.

* When the predation rate has decreased enough to be outpaced by the rabbit reproduction rate, the rabbit population reaches its minimum, and it begins to increase. From here, the cycle repeats.

In essence, the interplay between the fox and rabbit populations in this model is an elaborate negative feedback loop that causes the two populations to oscillate, not unlike a block on a spring or a simple pendulum.

Some Notes 1:


*   Rabbit population differential equation: (Rabbit \\(x\\) ; Fox \\(y\\) ):
\\[ \frac{{\rm d}x}{{\rm d}t} = \alpha x - \beta xy \\]
> The \\(\alpha x \\) term or "reproduction rate (term)" corresponds to the fact that the rabbit population growth depends on the population at a given time, which is a behavior shown by an exponential function. This fact becomes much clearer when we supress the second term: \\(\frac{dx}{dt}=\alpha x  x = A_{0}exp(\alpha t)\\), where \\(A_{0}\\) is the initial rabbit population. \
Implicitly, this should tell us that this model assumes that the rabbits would always have enough resources (food, water and such) in order to sustain their population ad infinitum in the absence of foxes (predators), making their current number the only factor affecting their reproduction rate barring the presence of foxes. Hence, we can say that \\(\alpha\\) controls how fast the rabbits reproduce at a given population size. \
The second term \\(-\beta xy\\) corresponds to the "death rate (term)" of the rabbit population (in this case, we can also think of it as the "predation rate (term)" since that's the only cause of death for rabbits that this  model considers). We can more clearly see this when we look at a specific rabbit population size \\(x\\) at time \\(t\\) and let fox population size \\(y\\) vary. Given that \\(\alpha , x , \beta , y\\) are all positive, increasing \\(y\\) means that the rabbit population growth rate decreases, and decreasing \\(y\\) means that the rabbit growth rate increases. This mirrors the fact that a greater number of foxes means that more predation could occur. \\(x\\) and \\(y\\) multiplied to one another alludes to the fact that for a given \\(y\\), how much rabbits the foxes consume also depend on the amount of rabbits there are. This tells us that \\(\beta\\) determines the rate predation for a given fox population size.   

* Fox population differential equation: (Rabbit \\(x\\) ; Fox \\(y\\) ):
\\[ \frac{{\rm d}y}{{\rm d}t} = \gamma xy - \delta y \\]
> Similar to the rabbit population growth equation, we can think of \\(\gamma xy\\) as the "reproduction rate (term)" of the fox population. Unlike the rabbit reproduction rate term, the fox reproduction rate term has the rabbit population as one of its factors aside from just a constant. This tells us that in this model, the reproduction rate of the fox population depends only on both how many foxes there are and how much rabbits are available for predation. This means that the factor \\(\gamma\\) determines how much the fox population can reproduce for a given rabbit population size. Here, the model implicitly assumes that the other needs of the foxes for survival are satisfied, and only the amount of rabbits (prey) available affect the reproduction rate term. \
The \\(-\delta y\\) term corresponds to the "death rate (term)" of the fox population. In this case, we assume that the only thing causing foxes to die is old age. If we assume that the life expectancy of each fox is the same throughout the population and unchanging in time, it then stands to reason that the death rate due to aging at a particular time is proportional to the fox population at that given time. \\(\delta\\) is this proportionality constant.  

* Some additional terms we can add to the model:
>If, for example, the rabbit population or fox population were stricken with a deadly disease, we could add another "death rate term" given by \\(-\kappa_1 x\\) for rabbits and \\(-\kappa_2 y\\) for foxes. Here \\(\kappa_1\\) and \\(\kappa_2\\) represent how deadly the disease is or the rate at which it kills rabbits or foxes.
