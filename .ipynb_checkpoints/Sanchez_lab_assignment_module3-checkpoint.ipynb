### PROBLEM 1
**Fourier transform of a simple function**

1. Calculate the coefficients in the discrete Fourier transforms of the modulated sine wave $y_n = \sin(\pi n/N) sin(20\pi n/N)$ sampled at $N = 1000$ evenly spaced points using your own user-defined function. *(15 pts.)*

2. Make a plot of their amplitudes similar to the plot shown in Fig. 7.4 of the Newman text. *(15 pts.)*


````python
# Importing necessary packages
import numpy as np
import matplotlib.pyplot as plt

# Defining y_n to be used
def y_n(_n,_N=1000, **kwargs):
  return (np.sin((np.pi*_n)/(_N)))*(np.sin((20*np.pi*_n)/(_N)))

# Defining DFT function to get standard quantity c_k of y_n
def DFT(N_=1000, _N_=1000, use_print=False, **kwargs):
  """
  Solves for the Fourier coefficients (standard quantity) of function y. This
  method is based on using trapezoidal integration to solve for the coefficients.

  N_ determines how many sample points are used
  _N_ is a parameter of y_n
  """
  # Defining upper limit of Fourier transform. We use int() to force it to be
  #an integer so that our array later can accept this as input
  upper_limit = int(np.floor((N_/2)+1))

  # Initializing the array c which will contain the standard quantities c_k
  c = np.zeros(upper_limit, complex)
  y_array = np.empty(N_,float)

  # range goes from 0 to N_-1 inclusive, which is all we'll need for
  #the DFT function
  for n in range(N_):
    y_array[n] = y_n(_n=n, _N=_N_)


  # Here, we run our loop to calculate for each c_k runing from k=0 to the upper
  #limit. (more on that in the notes below)
  for k in range (upper_limit):
    # We're doing the Fourier transform here for each c_k
    for n in range (N_):
      c[k] += y_array[n]*(np.exp(((-2j*np.pi*k*n)/(N_))))


  if use_print == True:
    return print(f"Each standard quantity c_k of y_n is given by:\n{c}")

  else:
    return c

DFT(use_print=True)

y_n_DFT = DFT()
abs_y_n_DFT = abs(DFT())

print(f"\nThe |c_k| values are given by: \n{abs_y_n_DFT}")
#----------------------------------------------------------------------------------------------------------------------------

# Plotting amplitudes
def Fourier_amp_plot(start_point=0, N=1000, use_xlimit=False, xlim1=-1, xlim2=40, title="Plot of |c_k|", **kwargs):
  """
  Plots the norm of the Fourier coefficients |c_k| on the y-axis and the corresponding k on
  the x-axis. The k values run from start_point to a un upper limit coresponding to N

  This funciton takes N sample points on the function for its Fourier transform.
  use_xlimit toggles whether or not plt.xlim() would be used, where xlim1 and
  xlim2 are its parameters.
  """
  upper_limit = int(np.floor((N/2)+1))

  # We set step to 1 since k must be an integer
  k_values = np.arange(start=start_point, stop=upper_limit, step=1)
  c_values = abs(DFT(N_=N))

  fig, ax = plt.subplots()
  ax.plot(k_values,c_values, marker="o", linestyle="--")

  plt.title(title)

  ax.set_xlabel("k values")
  ax.set_ylabel("|c_k| values")

  if use_xlimit == True:
    plt.xlim(xlim1, xlim2)


def plot_function(start_point=0, end_point=1000, precision=1000, xlabel="x", ylabel="y", title="", use_xlimit=False, xlim1=0, xlim2=100, marker="o", markersize=5, linestyle="--", **kwargs):
  """
  Plots specified function. precision determines how many points of the function
  are plotted. The points are evenly spaced, and the step size is automatically
  decided based on start_point and end_point.

  start_point is inclusive.
  end_point is inclusive.
  xlabel and ylabel sets the labels of the respective axes.
  use_xlimit toggles whether or not plt.xlim() would be used, where xlim1 and
  xlim2 are its parameters.
  """

  # Initializing the arrays to be used for the x-axis and the y-axis. Here, the
  #x-axis corresponds to the n values, and the y-axis corresponds to y_n
  x_values = np.linspace(start=start_point, stop=end_point, num=precision)
  y_values = np.empty(len(x_values),float)

  for i in range(len(x_values)):
    y_values[i] = y_n(_n=x_values[i])

  fig, ax = plt.subplots()
  ax.plot(x_values,y_values, marker=marker, markersize=markersize, linestyle=linestyle)

  plt.title(title)
  ax.set_xlabel(xlabel)
  ax.set_ylabel(ylabel)

  if use_xlimit == True:
    plt.xlim(xlim1, xlim2)


# This is the plot of y_n from n=0 to n=1000, inclusive
plot_function(start_point=0, end_point=1000, precision=1000, xlabel="n", ylabel="y_n", title="Fig 1. y_n from n=0 to n=1000")

# This is the plot of y_n from n=0 to n=3000
plot_function(start_point=0, end_point=3000, precision=3000, xlabel="n", ylabel="y_n", title="Fig 2. y_n from n=0 to n=3000", markersize=3)

# This is the |c_k| plot of y_n from k=0 to k=1000
Fourier_amp_plot(use_xlimit=True, xlim1=0, xlim2=50, title="Fig 3. Plot of |c_k|")



plt.show()



# Prints k value and corresponding c_k value os spike
print(f"\nSpike at k={10}, yielding c_k = {abs(y_n_DFT)[10]}")

````

Some additional notes#1:


*   upper_limit:
    >Since \\(\tt y_n\\) is real, we expect the complex components in its Fourier expansion to just cancel out, that is, there are terms that are complex conjugates of one another. Specifically, the terms from \\( n = \frac{N\_}{2} + 1 \\) until \\( n = N\_-1 \\) are complex conjugates of the terms runing from \\( n = \frac{N\_}{2} \\) to \\( n = 0 \\). This means that we only need to calculate about half of the coefficients given that the rest just repeat.

    >For even \\(N\_\\) , we only need: \\( k = \frac{N\_}{2} + 1 \\) coefficients

    >For odd \\(N\_\\) , we only need: \\( k = \frac{N\_+1}{2} \\) coefficients

    >Both cases can be conveniently covered by \\(  \frac{N\_}{2} +1 \\) or \\(  \frac{N\_ + 1}{2}  \\). In the code, the former was chosen.

    >If \\(N\\) is odd:
    >>For some non-negative integer m,

    >> \\( N\_ = 2m + 1  \frac{N\_ - 1}{2}  + 1 = m + 1  m + 1 = \frac{N\_ + 1}{2}\\) (This also says that \\( \frac{N\_ + 1}{2} \\) is an integer given that it's equal to one)

    >>\\(  \frac{N\_}{2} +1 =  \frac{2m+1}{2} +1  =  m + \frac{1}{2} +1 \\)

    >> Since we're taking the floor, we neglect \\(\frac{1}{2}\\):

    >>\\(  \frac{N\_}{2} +1 =  m + 1 =  m + 1 = \frac{N\_ + 1}{2}\\)

    >If \\(N\\) is even:
    >>For some non-negative integer m,

    >>\\( N\_ = 2m  \frac{N\_}{2} = m   m + 1 = \frac{N\_}{2} + 1\\)

    >>\\(  \frac{N\_}{2} +1 =  \frac{2m}{2} +1  =  m + 1 = m + 1 = \frac{N\_}{2} + 1 \\)



*   More on \\( y_n = sin(\frac{\pi n}{N})sin(\frac{20\pi n}{N}) = \frac{1}{2}cos(\frac{19\pi n}{N}) - \frac{1}{2}cos(\frac{21\pi n}{N})\\)
    > \\( cos(a)cos(b) - sin(a)sin(b) = cos(a+b)\\)
    > \\( cos(a)cos(b) + sin(a)sin(b) = cos(a-b)\\)

````python
# Importing necessary packages
import matplotlib.pyplot as plt

# Doing Fourier transform on dow
c_k_dow = np.fft.rfft(dow)

# Defining function that prunes the amplitudes
def keep_new_dow(c_k_array=c_k_dow, percent_kept=10, use_irfft=True, **kwargs):
  """
  This function takes a Fourier coefficient array c_k_array. It keeps the first
  percent_kept% of the values of the c_k_array and zeroes out the rest. After
  which, if use_irfft is True, it takes the inverse Fourier transform of the
  pruned array. Otherwise, it returns the pruned c_k_array.
  """
  # Initializing array that will contain 10% of the coefficients with the rest being
  #zeroed out
  kept_c_k_dow = np.zeros(len(c_k_array),float)

  # This is the index we'll be using to keep 10% of the coefficients. np.ceil()
  #ensures that we get 10% of the values, albeit we can overshoot a little, as is
  #the case here
  keep = int(np.ceil(len(c_k_array) *percent_kept/100))

  # We set kept_c_k_dow to contain the first 10% of the values
  for i in range(keep):
    kept_c_k_dow[i] = c_k_dow[i]

  if use_irfft == True:
    # Taking the inverse Fourier transform of kept_c_k_dow
    return np.fft.irfft(kept_c_k_dow)
  else:
    return kept_c_k_dow

# Contains the pruned c_k array
c_k_dow10 = keep_new_dow(percent_kept=10, use_irfft=False)

# Contains the inverse Fourier transform of the pruned c_k array
new_dow10 = keep_new_dow(percent_kept=10)

#-----------------------------------------------------------------------------------------------
# Plotting the two functions
fig, ax_dow = plt.subplots()

# Initializing days axis. We add 1 to len(dow) since np.arange doesn't include the endpoint
days = np.arange(start=1, stop=len(dow)+1, step=1)

# We add a 'label' argument to tell the legend which curve is which later
ax_dow.plot(days, dow, color="blue", label="Original")
ax_dow.set_xlabel("Day")
ax_dow.set_ylabel("Daily Closing Value")
ax_dow.tick_params("y", colors="blue")
plt.ylim(5000,15000)


def plot_new_dow_percent(axes, y, x=days, color="red", label="",**kwargs):
# We do this to make sure that the two graphs have the same scale

  axes.plot(x, y, color=color, label=label)
  axes.tick_params("y", colors=color)
  plt.ylim(5000,15000)

ax_new_dow10 = ax_dow.twinx()
plot_new_dow_percent(axes=ax_new_dow10, x=days, y=new_dow10, label= "1st 10% of c_k kept", color="green")

fig.legend(["Original", "1st 10% of c_k kept"])
plt.title(" Fig 4. Daily closing values vs Day")

plt.grid()

# We reuse a bit of code from the previous problem here
def fft_amp_plot(y, use_xlimit=False, xlim1=-1, xlim2=100, title="Plot of |c_k|", marker="o", markersize=5, linestyle="--", **kwargs):
  """
  Plots the norm of the Fourier coefficients |c_k| on the y-axis and the corresponding k on
  the x-axis. The k values run from zero

  use_xlimit toggles whether or not plt.xlim() would be used, where xlim1 and
  xlim2 are its parameters.
  """

  c_values = abs(y)
  k_values = np.arange(start=0, stop=len(c_values), step=1)

  fig, ax = plt.subplots()
  ax.plot(k_values,c_values, marker=marker, linestyle=linestyle, markersize=markersize)

  plt.title(title)

  ax.set_xlabel("k values")
  ax.set_ylabel("|c_k| values")

  if use_xlimit == True:
    plt.xlim(xlim1, xlim2)

fft_amp_plot(y=c_k_dow, title="Fig 5. |c_k| vs k", use_xlimit=True)
fft_amp_plot(y=c_k_dow10, title="Fig 6. |c_k| vs k keeping only 1st 10%", use_xlimit=True)

plt.show()
````

As shown, zeroing out the last 90% of the Fourier coefficients leads to a less noisy graph while keeping the most of the overall shape (the blue graph is the original data, while the green one is the pruned one). Given that most of the amplitudes one may find in a data set such as this one is noise that occurs at higher frequencies (higher k values), zeroing out most amplitudes "cleans" and smoothens the data, and makes the overarching trends more visible. We can also see this when comparing the unpruned \\(|c_k|\\) plot in Fig 5 with the pruned one in Fig 6.

While the trends become more visible, the values of the function seem to have been slightly distorted. This is not unexpected given that the smoothening process flattens some of the jagged and sharp peaks of the original function.

As such, the pruning makes the resulting graph more "derivative-friendly" and more wieldly in trying to predict how the dependent variable changes with a coresponding change in the independent variable because the pruning preserves most of the shape of the data. On the other hand, the afforementioned distortion suggests that using the pruned graph to predict the closing value for a particular day may not yield reliable results.
